## What is strings?
# - A string is a sequesce of characters used to store text in python.
### - Characters can be-
### - Letters- ("Python")
### - Numbers - "123"(still text,not math)
### - Symbols - "@#$"
### - Spaces- "Hello World"

```python
# Creating strings
#1. single quotes
text = 'Python'
print(text)
```

```python
# 2. double quotes
text = "Python"
print(text)
```

```python
# 3. triple quotes
text = """Python
is 
easy"""
print(text)
```

## Access Characters

```python
text = "Python"
print(text[0])
print(text[1])
print(text[5])
```

```python
# Negative indexing
text = "Python"
print(text[-1])
print(text[-2])
```

```python
# String Length
text = "Python"
print(len(text))
```

```python
# String Concatenation( joining strings together)
first = "Munira"
second = "Shumi"
print(first + second)
```

```python
# With spaces
text1 ="Munira" 
text2 = "Shumi"
print( text1 + " "+ text2)
```

```python
# Strings Repetation
text = "Hi"
print( text * 3 )
```

```python
# Common Strings Methods
# 1. upper()
text = "python"
print(text.upper())
```

```python
# 2. lower()
text = "python"
print(text.lower())
```

```python
# 3. title()
text = "python"
print(text.title())
```

```python
# 4. strip() -removes spaces from beginning and end.
text = " python "
print(text.strip())
```

```python
# 5. Replace()
text = " I Like Java"
print(text.replace("Java", "Python"))
```

```python
# 6. Split() - Converts strings into list.
text = " Apple banana mango"
print(text.split())
```

```python
# Checking Strings - in operator
text = " python "
print( "py" in text)
```

```python
text = " python "
print( "np" in text)
```
